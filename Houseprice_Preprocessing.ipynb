```python
# Libraries

import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("Train_HP.csv")
test = pd.read_csv("Test_HP.csv")
```

```python
train.head()
```

```python
train["source"] = "train"
test["source"] = "test"
```

```python
df = pd.concat([train,test])
df
```

```python
df.rename(columns={"RERA": "AREA"})
```

```python
df=df.rename(columns={"RERA": "AREA"})
```

```python
df_number_columns=df.select_dtypes(include = np.number).columns
```

```python
df_category_columns = df.select_dtypes(exclude = np.number).columns
```

```python
data_to_encode = df[df_category_columns].drop(columns = ["source"])
df_category_encoded = pd.get_dummies(data_to_encode)
```

```python
df_category_encoded
```

```python
df_final = pd.concat([df_category_encoded, df[df_number_columns],df["source"]], axis= 1)
```

```python
df_final
```

```python
train_preprocessed = df_final[df_final["source"]=="train"]
test_preprocessed = df_final[df_final["source"]=="test"]
```

```python
del train_preprocessed["source"]
del test_preprocessed["source"]
```

```python
train_preprocessed.to_csv("train_HPPSales_Preprocessed.csv", index = False)
test_preprocessed.to_csv("test_HPPSales_Preprocessed.csv", index = False)
```
